```python
# h
import matplotlib.pyplot as plt

role_counts = df_ml['role'].value_counts()

role_percentages = role_counts / role_counts.sum() * 100

plt.figure(figsize=(8, 8))
plt.pie(role_percentages, labels=role_percentages.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Heroes Based on Their Role')
plt.axis('equal')
plt.show()
```

7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
role_encoder = LabelEncoder()
df_ml['role'] = role_encoder.fit_transform(df_ml['role'])

# Separate features and target variable
X = df_ml.drop(['hero_name','role','win_rate','pick_rate','ban_rate','release_year',], axis=1)
y = df_ml['role']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the model
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: ", accuracy)

# Predict the role for the given character
new_character = {
  'defense_overall' :  6.0,
  'offense_overall' : 7.0,
  'skill_effect_overall' : 4.0,
  'difficulty_overall' : 3.0,
  'movement_spd' : 250,
  'magic_defense' : 9,
  'mana': 380,
  'hp_regen' : 37,
  'physical_atk' : 119,
  'physical_defense' : 24,
  'hp' : 2689,
  'attack_speed' :  0.75,
  'mana_regen' : 16.0,
}

new_character_df = pd.DataFrame([new_character])
new_character_df = new_character_df[X.columns]
predicted_role = clf.predict(new_character_df)

# Get the predicted role (dicoding)
predicted_role_name = role_encoder.inverse_transform(predicted_role)
```
